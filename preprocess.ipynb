```python
import numpy as np
import pandas as pd
```

```python
mnist_train = pd.read_csv('./mnist_train.csv', header=0)
mnist_train.describe()
```

```python
mnist_test = pd.read_csv('./mnist_test.csv', header=0)
mnist_test.describe()
```

```python
#function do divide datasets for binary classification
def binaryDataSet(i, j, dataSet):
    '''
        i, j are integers in the range [0, 9]. This function filters the dataset
        into a dataset containing only digits i, j.
    '''
    return dataSet.loc[dataSet['label'].isin([i, j])]
```

```python
#split training data
def split_mnist_training():
    for i in range(0, 9):
        for j in range(i + 1, 10):
            path = './mnist_binary/train/mnist_' + str(i) + '_' + str(j) + '.csv'
            binaryDataSet(i, j, mnist_train).to_csv(path)
    return 0

split_mnist_training()

#split test data
def split_mnist_test():
    for i in range(0, 9):
        for j in range(i + 1, 10):
            path = './mnist_binary/test/mnist_' + str(i) + '_' + str(j) + '.csv'
            binaryDataSet(i, j, mnist_test).to_csv(path)
    return 0

split_mnist_test()
```
